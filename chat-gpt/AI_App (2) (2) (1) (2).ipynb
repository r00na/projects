# Binary Search:(in Arabic)

```python
def binary_search(word, document):
    # تحويل النص إلى قائمة من الكلمات
    words = document.split()

    # تحديد حدود البحث
    low, high = 0, len(words) - 1

    while low <= high:
        mid = (low + high) // 2
        mid_word = words[mid]

        # إذا كانت الكلمة موجودة في الموقع المتوسط
        if mid_word == word:
            return True
        # إذا كانت الكلمة تأتي قبل الكلمة في الموقع المتوسط
        elif mid_word > word:
            high = mid - 1
        # إذا كانت الكلمة تأتي بعد الكلمة في الموقع المتوسط
        else:
            low = mid + 1

    # إذا لم تتم العثور على الكلمة
    return False

if __name__ == "__main__":
    # مثال على كيفية استخدام البرنامج
    document = "هذا هو مستند يستخدم للاختبار"

    search_word = "مستند"
    is_word_found = binary_search(search_word, document)

    if is_word_found:
        print(f"الكلمة '{search_word}' تم العثور عليها في المستند.")
    else:
        print(f"الكلمة '{search_word}' لم يتم العثور عليها في المستند.")
```
